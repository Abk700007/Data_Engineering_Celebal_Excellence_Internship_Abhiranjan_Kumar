# Load a CSV dataset into a Pandas DataFrame. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("combined_dataset.csv")
```

```python
# Converting Price realted columns to numeric
df['final_price'] = (
    df['final_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
```

```python
print(df['final_price'].dtype)

df[['initial_price','final_price']].head()
```

# Handle missing values (identify, fill/drop)


```python
df.isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

# Perform basic operations (filter rows, select columns)


```python
df[['product_id', 'title', 'product_description', 'rating',]].head()
```

```python
filtered_df = df[df['initial_price'] > 1000]

filtered_df.head()
```

# Remove duplicates

```python
print("Before:", len(df))

df = df.drop_duplicates()

print("After:", len(df))
```

# Create a derived column (total_amount = price * quantity)

```python
df['quantity'] = 10

df['total_amount'] = df['initial_price'] * df['quantity']
```

```python
df[['initial_price','quantity','total_amount']].head()
```

```python
df['popularity_score'] = (
    df['rating'] * df['ratings_count']
)

df[['rating',
    'ratings_count',
    'popularity_score']].head()
```

```python
df['popularity_score'].describe()
```

# ANALYSIS


## 1. Univariate Analysis
### Univariate analysis is performed to understand the distribution and characteristics of individual variables.

### Rating Distribution

```python
df['rating'].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['rating'], bins=20)

plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Products")

plt.show()
```

### Observations : -

- The average product rating is **3.62**.
- The median rating is **4.1**, indicating that most products are rated positively.
- A large concentration of products falls between **3.5 and 4.5 ratings**.
- A noticeable number of products have a rating of **0**, which may represent unrated or newly listed products.
- The maximum rating observed is **5.0**.

### Initial Price Distribution

```python
# Summary statistics for initial price

df['initial_price'].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['initial_price'], bins=30)

plt.title("Distribution of Initial Product Prices")
plt.xlabel("Initial Price")
plt.ylabel("Number of Products")

plt.show()
```

### Observations : -

- The average initial product price is approximately **₹2723**.
- The median price is **₹1999**, indicating that half of the products are priced below ₹1999.
- Most products are concentrated in the lower price range, primarily below **₹5000**.
- The distribution is right-skewed, with a small number of very high-priced products.
- The maximum product price observed is **₹22199**, suggesting the presence of premium products in the dataset.

## 2. Bivariate Analysis

### Bivariate analysis is used to understand the relationship between two variables.

### Rating vs popularity_score

```python
df[['rating',
    'ratings_count',
    'popularity_score']].describe()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['rating'],
    df['popularity_score'],
    alpha=0.6
)

plt.title("Rating vs Popularity Score")
plt.xlabel("Rating")
plt.ylabel("Popularity Score")

plt.show()
```

### Observations :-

- Most products have ratings between 3.5 and 4.5.
- Products with higher ratings generally tend to have higher popularity scores.
- A few products have exceptionally high popularity scores, appearing as outliers in the dataset.
- Products with ratings below 3.0 generally exhibit low popularity scores.
- The relationship is positive, but popularity is also influenced by the number of ratings received.

### Intial Price vs rating

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['initial_price'],
    df['rating'],
    alpha=0.6
)

plt.title("Initial Price vs Rating")
plt.xlabel("Initial Price")
plt.ylabel("Rating")

plt.show()
```

### Observation

- Most products are concentrated in the lower price range (below ₹5000).
- Products across different price ranges generally receive ratings between 3.5 and 4.5.
- No strong relationship is observed between product price and customer rating.
- High-priced products do not consistently receive higher ratings than lower-priced products.
- A few products have ratings of 0, which may indicate unrated or newly listed products.

## 3. Category-Level Analysis

## Category-level analysis helps identify which product categories contribute the most products in the dataset.

```python
# Top 10 product categories

df['category'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))

df['category'].value_counts().head(10).plot(
    kind='bar'
)

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.xticks(rotation=45)

plt.show()
```

### Observations : -

- The **tops** category contains the highest number of products (122), followed by **dresses** (100) and **shirts** (97).
- Clothing-related categories dominate the dataset, indicating a strong focus on apparel products.
- Categories such as **jackets** and **casual-shoes** have comparatively fewer products.
- The product distribution is not uniform across categories, with a few categories contributing a significant share of the total products.
